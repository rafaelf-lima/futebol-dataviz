```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("brasileirao.csv")
```

```python
df = df.rename(columns={"season": "temporada", "place": "posicao", "acronym": "acronimo", "team": "time",
                   "points": "pontos", "played": "jogos", "won": "vitorias", "draw": "empates", "loss": "derrotas",
                  "goals_for": "gols_pro", "goals_against": "gols_contra", "goals_diff": "diferenca_gols"}
          )
```

```python
df['time'] = df['time'].str.replace('SaoCaetano', 'São Caetano')
df['time'] = df['time'].str.replace('SaoPaulo', 'São Paulo')
df['time'] = df['time'].str.replace('RBBragantino', 'Red Bull Bragantino')
df['time'] = df['time'].str.replace('PontePreta', 'Ponte Preta')
df['time'] = df['time'].str.replace('SantoAndre', 'Santo André')
```

```python
cor_fundo = '#FFDEAD'
cor_barras = '#000000'

participacoes = df.groupby('time')['temporada'].count()
ax = participacoes.plot(kind='bar', color=cor_barras, figsize=(10, 6), edgecolor='#6495ED', linewidth=0.5)
plt.xlabel('Equipe')
plt.ylabel('Participações')
plt.title('Quantidade de participações por equipe')
plt.xticks(rotation=90)
for i, count in enumerate(participacoes):
    ax.annotate(str(count), xy=(i, count), ha='center', va='bottom', fontsize = 7)
ax.set_facecolor(cor_fundo)
plt.gcf().set_facecolor(cor_fundo)
plt.show()
```

```python
cor_fundo = '#FFDEAD'
cor_barras = '#000000'

participacoes = df.groupby('temporada')['gols_pro'].sum()
ax = participacoes.plot(kind='bar', color=cor_barras, figsize=(10, 6), edgecolor='#6495ED', linewidth=0.5)
plt.xlabel('Ano')
plt.ylabel('Gols')
plt.title('Quantidade de gols por ano')
plt.xticks(rotation=90)
for i, count in enumerate(participacoes):
    ax.annotate(str(count), xy=(i, count), ha='center', va='bottom', fontsize = 7)
ax.set_facecolor(cor_fundo)
plt.gcf().set_facecolor(cor_fundo)
plt.show()
```
